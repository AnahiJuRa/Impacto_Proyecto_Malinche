# 3.1 Recarga hídrica y regulación del agua

La Malinche actúa como un "regulador hídrico natural". Su importancia no radica solo en la topografía, sino en la estructura de sus suelos volcánicos y bosques templados.

## A. Funciones hidrológicas principales

### 1. Intercepción del dosel
Los árboles reducen la energía cinética de la lluvia, evitando el impacto directo sobre el suelo. Esto:
- minimiza la erosión inicial,
- favorece la infiltración,
- modera la entrada de agua a los mantos superiores.

### 2. Infiltración profunda en suelos volcánicos
Los suelos volcánicos de La Malinche presentan:
- alta macroporosidad,
- estructura granular,
- alta conductividad hidráulica,
- gran capacidad de almacenamiento.

Estas características facilitan que la lluvia se infiltre rápidamente hacia capas profundas.

### 3. Recarga de acuíferos
La infiltración profunda alimenta:
- acuíferos regionales,
- manantiales,
- sistemas de agua comunitaria.

Sin esta recarga, los acuíferos entran en estrés hídrico.

### 4. Regulación del flujo superficial
Los bosques:
- reducen el volumen de escorrentía,
- estabilizan caudales,
- disminuyen picos de inundación,
- evitan arrastre de sedimentos.

### 5. Mantenimiento de calidad del agua
El suelo volcánico funciona como filtro natural:
- retiene sedimentos,
- reduce patógenos,
- depura contaminantes.



## B. Riesgos hidrológicos ante la tala o construcción

Estudios en bosques de alta montaña muestran:

- Infiltración ↓ 25 a 60%
- Escorrentía ↑ 100 a 300%
- Sedimentos ↑ 200%
- Riesgo de deslaves ↑ 150%
- Pérdida de recarga de acuíferos acumulada cada año

La tala altera el equilibrio hidrológico y puede desencadenar:
- inundaciones repentinas,
- colapso de taludes,
- pérdida de manantiales,
- menor disponibilidad de agua potable.



# 3.2 Control de erosión y estabilidad del suelo (Versión extendida)

El control de erosión es uno de los servicios más críticos en montañas volcánicas, donde el suelo es fértil pero frágil.

## A. Funciones ecológicas del control de erosión

### 1. Estabilidad mecánica
Las raíces de árboles:
- anclan el suelo,
- distribuyen tensiones,
- previenen deslizamientos.

### 2. Reducción de erosión hídrica
La vegetación:
- dispersa la energía de la lluvia,
- genera rugosidad superficial,
- mantiene la integridad del horizonte orgánico.

### 3. Retención de sedimentos
El bosque funciona como barrera natural, evitando que sedimentos lleguen a cuerpos de agua.



## B. Impactos por pérdida de cobertura
Si se elimina la vegetación:
- El suelo pierde cohesión.
- Aumenta la erosión laminar, en surcos y cárcavas.
- Se forman cunetas que pueden evolucionar a barrancas.
- Aumenta el riesgo de deslizamientos por saturación.

Los suelos volcánicos pueden perder entre:
- 20 a 40 ton/ha/año en pendientes moderadas,
- más en zonas de intervención intensa.

La recuperación natural del suelo puede tardar entre:
- 60 y 200 años, dependiendo del clima y pendiente.

# 4. Interpretación de las ecuaciones utilizadas en el análisis ecológico

Las ecuaciones empleadas permiten cuantificar procesos ecológicos e hidrológicos fundamentales para entender cómo la intervención propuesta afectaría al ecosistema de La Malinche. A continuación se presentan sus interpretaciones científicas y las conclusiones que se derivan de cada una.



# 4.1 Ecuación de Infiltración Hídrica

$$
I = P - (Q + ET)
$$



Esta ecuación describe el **balance hídrico** del ecosistema. La infiltración (**I**) es el agua que realmente entra al suelo después de una lluvia.

- **P** = precipitación total  
- **Q** = escorrentía (agua superficial que no se infiltra)  
- **ET** = evapotranspiración (evaporación + transpiración de plantas)

### Interpretación ecológica

- En un **bosque sano**, la escorrentía es baja y la infiltración es alta.  
- Con **tala o compactación**, la escorrentía aumenta y la infiltración disminuye.

### Conclusión

> La tala reduce drásticamente la infiltración, disminuyendo la recarga de acuíferos y aumentando las inundaciones y la erosión. Esto afecta el suministro de agua para Puebla y Tlaxcala.



# 4.2 Ecuación Universal de Pérdida de Suelo (USLE)

$$
A = R \cdot K \cdot LS \cdot C \cdot P
$$

### ¿Qué significa?

Estima la cantidad de suelo que se pierde por erosión cada año (**A**, en ton/ha/año).

Las variables representan:

- **R** = erosividad de la lluvia  
- **K** = erodabilidad del suelo  
- **LS** = pendiente y longitud de la ladera  
- **C** = cobertura vegetal  
- **P** = prácticas de conservación  

### Interpretación ecológica

- En La Malinche, los suelos volcánicos (K alto) y las pendientes pronunciadas (LS alto) son **muy vulnerables** a la erosión.
- La pérdida de cobertura vegetal (aumento de C) multiplica la erosión entre 5 y 10 veces.

### Conclusión

> La intervención en zonas de pendiente provoca erosión severa, deslizamientos y pérdida de suelo fértil que puede tardar siglos en recuperarse. Uno de los impactos más graves e irreversibles.



# 4.3 Ecuación de Captura de Carbono

$$
C = \text{Biomasa} \times 0.47
$$

### ¿Qué significa?

Aproximadamente el **47% de la biomasa de un árbol es carbono**.  
Esta ecuación calcula cuánto carbono almacena una hectárea de bosque.

### Interpretación ecológica

Mientras los árboles crecen, capturan CO₂ de la atmósfera y lo fijan como carbono en su biomasa.

### Conclusión

> Cada hectárea talada libera grandes cantidades de carbono almacenado, debilitando el papel del bosque como sumidero climático.



# 4.4 Conversión de Carbono a CO₂

$$
CO_2 = C \times 3.67
$$

### ¿Qué significa?

La molécula de CO₂ pesa 3.67 veces más que un átomo de carbono, por lo que esta ecuación convierte el carbono almacenado en **emisiones reales de dióxido de carbono**.

### Interpretación ecológica

Si un área de bosque se tala, el carbono almacenado en la biomasa se libera como CO₂.

### Conclusión

> La pérdida de unas pocas hectáreas de bosque puede generar cientos de toneladas de CO₂, equivalentes a las emisiones anuales de cientos de personas.



#  Conclusión general

Las ecuaciones hidrológicas, de erosión y de carbono muestran que:

- La tala **colapsa la infiltración**, afectando la disponibilidad de agua.  
- La erosión se **multiplica**, comprometiendo estabilidad del suelo.  
- La captura de carbono **disminuye drásticamente**, agravando el cambio climático.  
- La pérdida de bosque implica **emisiones significativas** de CO₂.  

**Matemática y ecológicamente, el impacto acumulado del Ecoparque Tlalli–Malinche sería grave, acelerado y en gran parte irreversible.**




# 4. Impacto hidrológico por tala

La tala y fragmentación del bosque en La Malinche provoca alteraciones profundas en el ciclo hidrológico. Dado que el sistema depende de suelos volcánicos altamente porosos, cualquier intervención tiene repercusiones inmediatas en la infiltración, escorrentía y disponibilidad de agua para las comunidades de Puebla y Tlaxcala.

Este impacto es uno de los más graves y con efectos a mediano y largo plazo.



# 4.1 ¿Cómo funciona el ciclo hidrológico en un bosque sano?

Un bosque bien conservado cumple tres funciones fundamentales:

### **1. Infiltración profunda del agua**
El agua de lluvia penetra en:
- la hojarasca,
- el horizonte orgánico,
- el suelo volcánico de alta porosidad.

Esto alimenta directamente los **acuíferos subterráneos**.

### **2. Regulación de escorrentía**
El bosque reduce el flujo superficial mediante:
- raíces,
- troncos,
- microtopografía,
- absorción del suelo.

Evita inundaciones y controla el arrastre de sedimentos.

### **3. Intercepción del dosel**
Las copas de los árboles atrapan parte de la lluvia antes de tocar el suelo, reduciendo su energía y evitando erosión inicial.


# 4.2 Ecuación hidrológica fundamental

La infiltración se calcula con la ecuación:

$$
I = P - (Q + ET)
$$

Donde:

- **I** = infiltración  
- **P** = precipitación  
- **Q** = escorrentía  
- **ET** = evapotranspiración  

###  Interpretación:

- Si **Q aumenta** (más agua corriendo sobre la superficie)  
- y **ET disminuye** (menos árboles → menos transpiración)

Entonces:

$$
I \text{ disminuye significativamente}
$$



# 4.3 ¿Qué cambia cuando se tala el bosque?

### **1. Disminuye la infiltración**
Estudios en ecosistemas montanos muestran reducciones de:
- **25% a 60%** en la infiltración.

Los suelos volcánicos, al compactarse con maquinaria o perder su estructura humosa, pierden capacidad de absorber agua.

### **2. Aumenta la escorrentía**
La escorrentía puede incrementar entre:
- **100% a 300%**,  
dependiendo del grado de disturbio.

Esto significa:
- más erosión,
- más transporte de sedimentos,
- más riesgo de deslaves.

### **3. Menor recarga del acuífero**
La Malinche alimenta:
- manantiales en Canoa,
- corrientes subterráneas hacia Puebla,
- acuíferos agrícolas en Tlaxcala.

La reducción en recarga implica:
- menos agua potable,
- estrés hídrico acumulado,
- agotamiento del acuífero.

# 4.5 Ecuación de Balance Hídrico (Thornthwaite-Mather)

Otra forma de ver el impacto es mediante el almacenamiento de agua en el suelo:

$$
Almac_{t} = Almac_{t-1} + P - ET
$$

### Interpretación:
- Sin bosque → ET disminuye, pero al compactarse el suelo también **disminuye Almac**,  
- El sistema pierde capacidad de retención,  
- Más agua se pierde como escorrentía.



# 4.6 Riesgos específicos para La Malinche (hidrología)

### **1. Pérdida de manantiales**
Los manantiales dependen directamente de la infiltración profunda.  
Menos infiltración = menos aportes subterráneos.

### **2. Inundaciones repentinas**
La vegetación regula la velocidad del agua.  
Sin ella, el agua corre sin fricción →  
**mayor riesgo de deslaves e inundaciones** en Canoa y partes bajas.

### **3. Menor disponibilidad de agua potable**
Las comunidades dependen del acuífero recargado por el volcán.  
La tala podría provocar:
- reducción del volumen extraíble,  
- pozos secos o con niveles menores,  
- estrés hídrico crónico.

### **4. Contaminación del agua**
La erosión transporta:
- sedimentos,
- metales del suelo volcánico,
- contaminantes de la construcción.

Esto reduce la calidad del agua.



# 4.7 Conclusión de impacto hidrológico

> La intervención en La Malinche alteraría profundamente su función como regulador hídrico regional. La disminución de la infiltración y el incremento de la escorrentía comprometen la recarga de acuíferos y aumentan el riesgo de erosión, deslaves e inundaciones. Este impacto no solo afecta al ecosistema, sino también al suministro de agua potable y la seguridad ambiental de Puebla y Tlaxcala.

# 6. Erosión y estabilidad del suelo

La Malinche es una montaña de origen volcánico con pendientes fuertes y suelos fértiles pero frágiles.  
Cuando el bosque está sano, las raíces, la hojarasca y la estructura del suelo mantienen la estabilidad.  
Cuando se tala, se compacta o se construye infraestructura (caminos, estacionamientos, edificios), el suelo pierde esa protección y se vuelve muy vulnerable:

- a la **erosión** (el suelo se lo lleva el agua),  
- a los **deslizamientos** (el suelo se mueve ladera abajo),  
- a la **pérdida de fertilidad** (el suelo fértil superior desaparece).

Esta sección explica **cómo y por qué** pasa eso, usando modelos y ecuaciones, pero con interpretación clara.



## 6.1 ¿Qué es la erosión del suelo?

La **erosión** es el proceso mediante el cual el suelo es desprendido y transportado por agentes como:

- agua (lluvia, escorrentía),
- viento,
- gravedad (deslizamientos).

En La Malinche, el principal agente es el **agua de lluvia en laderas volcánicas**.

Cuando hay bosque:

- la lluvia cae sobre las copas,
- se reduce la energía de las gotas,
- la hojarasca y raíces protegen el suelo,
- el agua se infiltra en lugar de correr por la superficie.

Cuando se quita la vegetación:

- las gotas golpean directo el suelo,
- se forman chorros de agua que “peinan” la superficie,
- se llevan partículas finas cuesta abajo,
- se forman surcos, cárcavas y barrancas.



## 6.2 Ecuación Universal de Pérdida de Suelo (USLE)

Para estimar la erosión promedio anual se utiliza la USLE:

$$
A = R \cdot K \cdot LS \cdot C \cdot P
$$

Donde:

- **A** = pérdida de suelo (toneladas por hectárea por año)  
- **R** = erosividad de la lluvia  
- **K** = erodabilidad del suelo (qué tan fácil se desarma)  
- **LS** = factor de pendiente y longitud de ladera  
- **C** = factor de cobertura (si hay o no vegetación)  
- **P** = factor de prácticas de conservación (terrazas, obras, etc.)

### Interpretación sencilla

- Si llueve fuerte → **R** es grande.  
- Si el suelo es volcánico, suelto → **K** es grande.  
- Si la ladera es inclinada → **LS** es grande.  
- Si talas el bosque → **C** aumenta mucho.  
- Si no hay manejo de conservación → **P** es grande.

Resultado:  
> **A se dispara → se pierde muchísimo suelo cada año.**

En un bosque sano, el factor C es muy bajo (el suelo está cubierto y protegido).  
En una zona intervenida, C sube porque el suelo queda desnudo o con cemento.



## 6.3 ¿Por qué La Malinche es especialmente vulnerable?

La Malinche combina tres factores peligrosos:

1. **Suelos volcánicos**  
   - Muy fértiles pero sueltos.
   - Estructura porosa que se rompe si se compacta.

2. **Pendientes pronunciadas**  
   - La gravedad ayuda a que el suelo se deslice cuesta abajo.
   - Cualquier disturbio se amplifica por la inclinación.

3. **Lluvias intensas en temporada**  
   - Lluvias fuertes en poco tiempo generan mucha escorrentía si no se infiltran.

Por eso, aunque el proyecto “solo” toque decenas de hectáreas, el impacto puede extenderse ladera abajo, afectando:

- barrancas,
- caminos,
- parcelas,
- posiblemente viviendas y comunidades como Canoa.



## 6.4 Compactación del suelo por maquinaria y construcción

La construcción de un parque de este tipo implica:

- maquinaria pesada,
- nivelación de terreno,
- rellenos y cortes,
- pavimento, concreto, grava.

### ¿Qué hace la compactación?

- Aumenta la **densidad aparente** del suelo (se vuelve más “apretado”).  
- Disminuye la **porosidad** (menos espacio para aire y agua).  
- Disminuye la **infiltración** (el agua ya no entra, se escurre).  
- Rompe estructuras del suelo y canales de raíces.

Consecuencia directa:

> Menos infiltración → más escorrentía → más erosión.

Incluso si luego “reforestan”, un suelo muy compactado ya no se comporta igual que un bosque maduro.


## 6.5 Deslizamientos y estabilidad de laderas (modelo de talud infinito)

Para analizar si una ladera es estable o no, se usa un concepto llamado **Factor de Seguridad (FS)**.

Una versión simplificada del modelo de talud infinito para suelos saturados es:

$$
FS = \frac{c' + (\gamma \cdot z \cdot \cos^2 \theta - u)\tan \varphi'}{\gamma \cdot z \cdot \sin \theta \cdot \cos \theta}
$$

Donde:

- **FS** = factor de seguridad (si FS < 1, el talud tiende a fallar)  
- **c'** = cohesión efectiva del suelo  
- **γ** = peso unitario del suelo  
- **z** = profundidad del plano de falla (espesor de suelo)  
- **θ** = ángulo de la pendiente  
- **u** = presión de poros (agua en el suelo)  
- **φ'** = ángulo de fricción interna del suelo

### Lectura intuitiva

- El **numerador** representa fuerzas que resisten el deslizamiento:
  - cohesión del suelo,
  - fricción interna (efecto de las raíces, estructura, etc.).

- El **denominador** representa fuerzas que favorecen el deslizamiento:
  - peso del suelo sobre una pendiente inclinada.

Si:

- se **satura** el suelo de agua (lluvias intensas + menos infiltración),  
- se **pierden raíces** por tala,  
- se **modifica la pendiente** por cortes para caminos o terrazas,

entonces:

- la resistencia baja,
- la fuerza que empuja cuesta abajo aumenta,
- el **FS disminuye**.

Cuando **FS < 1** → la ladera es inestable y se favorecen **deslizamientos**.

## 6.7 Riesgos específicos para comunidades y ecosistemas

### 1. Formación de cárcavas y barrancas

El agua concentrada en caminos, zanjas o taludes:
- se lleva el suelo,
- profundiza surcos,
- destruye microhábitats,
- puede cortar veredas y accesos.

### 2. Deslizamientos locales

En pendientes con suelo suelto y saturado:
- pueden ocurrir deslizamientos de tierra,
- afectar caminos de acceso,
- comprometer infraestructura,
- representar riesgo para personas.

### 3. Aporte masivo de sedimentos

La erosión en la parte alta:
- llena barrancas con sedimento,
- afecta arroyos y posibles cauces intermitentes,
- impacta la calidad del agua (más turbiedad, menos calidad).

### 4. Pérdida de fertilidad

El suelo superficial (horizonte orgánico) es el más fértil y rico en materia orgánica.

Si se lo lleva el agua:
- la productividad ecológica baja,
- el bosque se regenera más lento o no se regenera,
- la restauración se vuelve más costosa y menos efectiva.

# 8. Base matemática del análisis

En esta sección se presentan las principales ecuaciones y modelos utilizados para describir y cuantificar los impactos ambientales asociados al proyecto en La Malinche.  
El objetivo no es solo mostrar fórmulas, sino explicar **qué representan**, **cómo se interpretan** y **qué conclusiones ecológicas permiten obtener**.

Se agrupan en cinco bloques:

- Hidrología (agua)
- Erosión del suelo
- Estabilidad de laderas
- Biodiversidad y fragmentación
- Carbono y cambio climático



## 8.1 Modelos hidrológicos

### 8.1.1 Balance hídrico simple

La ecuación básica del balance hídrico es:

$$
I = P - (Q + ET)
$$

Donde:

- \( I \) = infiltración
- \( P \) = precipitación (lluvia)
- \( Q \) = escorrentía superficial
- \( ET \) = evapotranspiración

**Interpretación:**

- En un bosque sano: \( Q \) es bajo, \( ET \) es moderada → \( I \) es alta.
- Con tala y suelo compactado: \( Q \) aumenta, \( ET \) disminuye → \( I \) baja.

**Conclusión ecológica:**

> Menos infiltración implica menor recarga de acuíferos, más agua corriendo en superficie y mayor riesgo de erosión, inundaciones y pérdida de suelo.



### 8.1.2 Modelo de Horton (infiltración en el tiempo)

La capacidad de infiltración del suelo suele disminuir durante una lluvia intensa.  
El modelo de Horton lo expresa así:

$$
f(t) = f_c + (f_0 - f_c)\, e^{-k t}
$$

Donde:

- \( f(t) \) = capacidad de infiltración en el tiempo \( t \)
- \( f_0 \) = capacidad de infiltración inicial (al inicio de la lluvia)
- \( f_c \) = capacidad de infiltración final (estado casi saturado)
- \( k \) = parámetro de decaimiento
- \( t \) = tiempo

**En un bosque sano:**

- \( f_0 \) y \( f_c \) son altos (suelo poroso, con raíces y hojarasca).

**En un suelo compactado:**

- \( f_0 \) y \( f_c \) son mucho más bajos (agua entra menos al suelo).

**Conclusión:**

> La compactación por maquinaria y pérdida de materia orgánica disminuyen tanto la infiltración inicial como la final, lo que incrementa escorrentía y erosión.



### 8.1.3 Balance de almacenamiento en el suelo (Thornthwaite-Mather)

Un modelo simplificado para el almacenamiento de agua en el suelo es:

$$
Almac_t = Almac_{t-1} + P - ET
$$

Donde:

- \( Almac_t \) = agua almacenada en el suelo en el tiempo \( t \)
- \( P \) = precipitación
- \( ET \) = evapotranspiración

En un sistema sano, el suelo puede aumentar o conservar su almacenamiento; en un sistema degradado, pierde capacidad de retención.



## 8.2 Modelos de erosión del suelo

### 8.2.1 USLE (Ecuación Universal de Pérdida de Suelo)

La USLE estima la pérdida media anual de suelo:

$$
A = R \cdot K \cdot LS \cdot C \cdot P
$$

Donde:

- \( A \) = pérdida de suelo (ton/ha/año)
- \( R \) = erosividad de la lluvia
- \( K \) = erodabilidad del suelo
- \( LS \) = longitud y pendiente de la ladera
- \( C \) = factor de cobertura (vegetación)
- \( P \) = prácticas de conservación

**Lectura ecológica:**

- En La Malinche, \( R \) (lluvias intensas), \( K \) (suelo volcánico) y \( LS \) (pendientes) ya son altos.
- Si aumentan \( C \) (suelo desnudo) y \( P \) (sin conservación), \( A \) crece de forma dramática.

**Conclusión:**

> Cambiar el uso de suelo en pendientes volcánicas multiplica la pérdida anual de suelo, generando daños en laderas, caminos, barrancas y cuerpos de agua.



## 8.3 Modelos de estabilidad de laderas

### 8.3.1 Factor de Seguridad (FS) – Modelo de talud infinito simplificado

El **Factor de Seguridad** compara fuerzas que resisten el deslizamiento con fuerzas que lo impulsan:

$$
FS = \frac{\text{Fuerzas resistentes}}{\text{Fuerzas deslizantes}}
$$

Una forma común en suelos saturados es:

$$
FS = \frac{c' + (\gamma \cdot z \cdot \cos^2 \theta - u)\tan \varphi'}{\gamma \cdot z \cdot \sin \theta \cdot \cos \theta}
$$

Donde:

- \( c' \) = cohesión efectiva del suelo
- \( \gamma \) = peso unitario del suelo
- \( z \) = profundidad del plano de falla
- \( \theta \) = ángulo de la pendiente
- \( u \) = presión de poros (agua dentro del suelo)
- \( \varphi' \) = ángulo de fricción interna

**Regla básica:**

- Si \( FS > 1.3 \): ladera estable.
- Si \( FS \approx 1 \): ladera en equilibrio límite.
- Si \( FS < 1 \): riesgo de deslizamiento.

**¿Qué hace la tala?**

- Disminuye \( c' \) y efectivamente la fricción (se pierden raíces).
- Aumenta \( u \) (más agua en el suelo por menor infiltración profunda y más saturación).
- El resultado es una disminución de \( FS \).

**Conclusión:**

> La tala y la compactación del suelo en pendientes pronunciadas como las de La Malinche incrementan el riesgo de deslizamientos y fallas de ladera.



## 8.4 Modelos de biodiversidad y fragmentación

### 8.4.1 Relación especie–área

Una relación clásica en ecología es:

$$
S = c \cdot A^z
$$

Donde:

- \( S \) = número de especies
- \( A \) = área de hábitat
- \( c, z \) = constantes empíricas

Cuando el área disponible disminuye, el número de especies tiende a reducirse.

**Conclusión:**

> Reducir el área de bosque continuo disminuye el número de especies que el ecosistema puede sostener, sobre todo las más sensibles o especializadas.



### 8.4.2 Índices de fragmentación y conectividad

Aunque aquí no se calculan con software específico, conceptualmente se manejan:

**Índice de fragmentación (PF):**

$$
PF = \frac{N_{fragmentos}}{A_{total}}
$$

- A mayor \( PF \): mayor fragmentación del paisaje.

**Índice de conectividad (CI):**

$$
CI = \frac{A_{\text{parches conectados}}}{A_{\text{total}}}
$$

- A menor \( CI \): mayor aislamiento, menor conectividad ecológica.

**Conclusión:**

> La construcción del Ecoparque genera nuevos fragmentos y bordes, reduciendo la conectividad ecológica y aumentando el aislamiento de poblaciones de flora y fauna.



## 8.5 Modelos de carbono y cambio climático

### 8.5.1 Estimación de carbono almacenado

Una aproximación común es:

$$
C = \text{Biomasa} \times 0.47
$$

Donde:

- \( C \) = carbono almacenado (ton)
- 0.47 ≈ fracción de carbono en la biomasa seca

### 8.5.2 Conversión de carbono a CO\(_2\)

Para obtener el CO\(_2\) equivalente:

$$
CO_2 = C \times 3.67
$$

**Ejemplo simple:**

Si en una hectárea hay 100 t de biomasa:

- \( C = 100 \times 0.47 = 47 \) t de C  
- \( CO_2 = 47 \times 3.67 \approx 172.5 \) t de CO\(_2\)

**Conclusión:**

> La tala de unas pocas hectáreas de bosque maduro puede liberar cientos de toneladas de CO\(_2\) a la atmósfera, convirtiendo a La Malinche de sumidero de carbono en fuente de emisiones.



## 8.6 Síntesis de la base matemática

Las ecuaciones anteriores permiten:

- **Cuantificar** cómo se reduce la infiltración y se incrementa la escorrentía.
- **Estimar** cuánta pérdida de suelo ocurre bajo distintos escenarios.
- **Evaluar** el riesgo de deslizamientos en función de la pendiente y la saturación.
- **Relacionar** la pérdida de hábitat con la pérdida de especies.
- **Traducir** la tala de bosque en emisiones de CO\(_2\).

En conjunto, la base matemática refuerza el argumento ecológico:

> No se trata solo de una cuestión estética o simbólica; las funciones físicas y biológicas de La Malinche están gobernadas por procesos cuantificables. Los modelos indican que la intervención propuesta empuja al sistema más allá de umbrales críticos en agua, suelo, biodiversidad y clima, generando daños acumulativos y de difícil reversión.

##  Conexión geotécnica: estabilidad de laderas y Factor de Seguridad

### Qué predice la ecuación

- En pendientes pronunciadas (θ alto) → el sistema es muy sensible.

### Relación con la gráfica
La gráfica del “Factor de Seguridad” lo representa como:

- Bosque sano: FS = 1.5 (estable)  
- Zona talada: FS = 0.9 (riesgo)

Exactamente lo que el modelo predice.


> Las raíces actúan como un “entramado” geotécnico. Si lo quitas, la montaña pierde resistencia. La gráfica hace visible algo que el FS cuantifica: la posibilidad real de deslaves o colapsos del suelo.

## Conexión climática: carbono, CO₂ y cambio climático

### Ecuaciones
1. Carbono almacenado:
   \[
   C = \text{Biomasa} \times 0.47
   \]

2. Conversión a CO₂:
   \[
   CO_2 = C \times 3.67
   \]

### Qué predicen
- La tala libera el carbono almacenado en la biomasa vegetal.  
- Cada tonelada de carbono equivale a 3.67 t de CO₂.  

### Relación con el ecosistema
Si una hectárea tiene 100 toneladas de biomasa:

- C = 47 t de carbono  
- CO₂ liberado = 172.5 t  




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.linspace(1, 100, 100)  # área en %
c = 10  # constante hipotética
z = 0.25

S = c * (A/100)**z

plt.figure(figsize=(7,4))
plt.plot(A, S)
plt.title("Relación especie–área en un bosque montano")
plt.xlabel("Porcentaje de área conservada (%)")
plt.ylabel("Número relativo de especies (S)")
plt.grid(True)
plt.show()
```

Si pierde 20% del área → se pierde entre 5–12% de las especies.

Si pierde 40% del área → pérdida de 15–20% de especies.

Aunque la curva parece “suave”, representa colapsos locales de biodiversidad.

```python
biomasa = np.linspace(20, 200, 50)
C = biomasa * 0.47
CO2 = C * 3.67

plt.figure(figsize=(7,4))
plt.plot(biomasa, CO2)
plt.title("CO₂ liberado según biomasa perdida")
plt.xlabel("Biomasa removida (t/ha)")
plt.ylabel("CO₂ equivalente liberado (t/ha)")
plt.grid(True)
plt.show()
```

Una hectárea con 100 toneladas de biomasa → libera ~172 t de CO₂.

Talando 10 hectáreas → 1,720 t de CO₂, equivalente a lo que emiten 400 automóviles en un año.

```python
pendiente = np.radians(np.linspace(10, 40, 30))  # grados → radianes
lluvia = np.linspace(0, 200, 30)  # mm/día (saturación)

FS = np.zeros((len(pendiente), len(lluvia)))

for i, theta in enumerate(pendiente):
    for j, rain in enumerate(lluvia):
        u = rain / 100  # presión de poros simplificada
        FS[i, j] = (5 + (18*np.cos(theta)**2 - u)*np.tan(np.radians(30))) / (18*np.sin(theta)*np.cos(theta))

plt.figure(figsize=(8, 6))
plt.imshow(FS, extent=[0,200,10,40], origin='lower', aspect='auto', cmap='coolwarm')
plt.colorbar(label="FS")
plt.xlabel("Lluvia (mm/día)")
plt.ylabel("Pendiente (°)")
plt.title("Simulación del Factor de Seguridad según lluvia y pendiente")
plt.show()
```

```python
FS < 1 = riesgo de deslave (zona roja).

El área roja crece si aumenta la lluvia, aumenta la pendiente, se pierde la vegetación.

Esta gráfica muestra por qué La Malinche es extremadamente vulnerable a deslizamientos si se talan zonas clave.
```

# 10 Conclusiones globales finales

El análisis matemático, hidrológico, ecológico y geomorfológico muestra con claridad que La Malinche no es un simple conjunto de árboles: es un sistema de soporte vital para Puebla y Tlaxcala. Cada ecuación, cada modelo y cada gráfica apunta hacia la misma conclusión:



##  10.1 Conclusión ecológica
La Malinche alberga más de 900 especies, muchas de ellas endémicas.  
Reducir el área del bosque implica reducir el número de especies, fracturar corredores biológicos y debilitar procesos ecológicos esenciales como:

- polinización,  
- dispersión de semillas,  
- regulación de plagas,  
- producción primaria.

Incluso pequeñas pérdidas de área provocan caídas desproporcionadas en biodiversidad.



## 10.2 Conclusión hidrológica
El volcán funciona como una “esponja natural” que alimenta acuíferos.

La tala o compactación:

- reduce la infiltración entre 25–60%,  
- aumenta la escorrentía 100–300%,  
- eleva el riesgo de inundaciones,  
- disminuye el agua disponible para manantiales y comunidades.

La gráfica de infiltración vs escorrentía lo vuelve evidente.



## 10.3 Conclusión sobre erosión y suelo
La USLE demuestra que la pérdida de suelo:

- se multiplica x5 a x10 en zonas taladas,  
- supera fácilmente las 30–40 t/ha/año.

La montaña pierde suelo fértil que no se recupera en décadas.

La pérdida de raíces aumenta el riesgo de deslizamientos, como confirmó la gráfica del **Factor de Seguridad**.



##  10.4 Conclusión climática
El bosque de La Malinche es un sumidero de carbono altamente eficiente.

Talando:

- se liberan cientos de toneladas de CO₂ por hectárea,  
- aumenta la temperatura local,  
- se reduce la humedad ambiental,  
- se favorecen incendios.

El bosque pasa de ser **protección climática** a **fuente de gases de efecto invernadero**.



##  10.5 Conclusión socioambiental
Las comunidades que dependen del agua, del suelo fértil y del clima estable serían afectadas.  
El paisaje cultural y espiritual también se degrada.

El desarrollo propuesto no compensa los daños ecológicos, hidrológicos y sociales.



##  13.6 Conclusión final integradora
> **El proyecto Ecoparque Tlalli–Malinche no es viable en ninguna dimensión científica.  
> La evidencia cuantitativa, las gráficas, las ecuaciones y los datos reales convergen en que la intervención produciría daños irreversibles en agua, suelo, biodiversidad, clima y estabilidad geomorfológica.**

La ciencia es clara:  
**La Malinche debe permanecer intacta.**
